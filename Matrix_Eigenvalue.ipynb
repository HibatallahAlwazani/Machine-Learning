```python
from PIL import Image
from numpy import linalg as LA

import matplotlib.pyplot as plt
import numpy as np

import matplotlib.image as mpimg
import cv2
```

```python
img = cv2.imread("cutiecat.jpg",flags=cv2.IMREAD_UNCHANGED)
img = cv2.cvtColor(img,cv2.COLOR_BGR2RGB)
rows,cols,_ = img.shape
imgplot = plt.imshow(img)
plt.show()
```

```python
M = np.float32([[2, -np.sin(0.3), 0],[0,  1*np.cos(0.3), 0]])
out = cv2.warpAffine(img,M,(cols,rows))
imgplot = plt.imshow(out)

plt.show()
```

```python
M = cv2.getRotationMatrix2D((cols/2,rows/2),45,1)
out = cv2.warpAffine(img,M,(cols,rows))
print(M)

imgplot = plt.imshow(out)
plt.show()
```

```python

pts1 = np.float32([[50,50],[0,0],[0,0]])
pts2 = np.float32([[10,100],[200,50],[100,250]])

M = cv2.getAffineTransform(pts1,pts2)

dst = cv2.warpAffine(img,M,(cols,rows))
imgplot = plt.imshow(dst)
plt.show()
```
